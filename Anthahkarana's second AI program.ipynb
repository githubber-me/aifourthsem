```python
from collections import deque

class State:
    def __init__(self, l_m, l_c, b_p):
        self.l_m = l_m
        self.l_c = l_c
        self.r_m = 3 - l_m
        self.r_c = 3 - l_c
        self.b_p = b_p

    def __eq__(self, other):
        return (self.l_m == other.l_m and self.l_c == other.l_c and self.b_p == other.b_p)

    def __hash__(self):
        return hash((self.l_m, self.l_c, self.b_p))

    def __str__(self):
        return f"({self.l_m}{self.l_c}:{self.r_m}{self.r_c})-{self.b_p}"

def is_valid(state):
    if state.l_m < 0 or state.l_c < 0 or state.r_m < 0 or state.r_c < 0 or \
            (state.l_m > 0 and state.l_m < state.l_c) or \
            (state.r_m > 0 and state.r_m < state.r_c):
        return False
    return True

def generate_next_states(current):
    next_states = []
    if current.b_p == 'left':
        for m in range(min(2, current.l_m) + 1):
            for c in range(min(2, current.l_c) + 1):
                if m + c <= 2 and m + c >= 1:
                    next_state = State(current.l_m - m, current.l_c - c, 'right')
                    if is_valid(next_state):
                        next_states.append(next_state)
    else:
        for m in range(min(2, current.r_m) + 1):
            for c in range(min(2, current.r_c) + 1):
                if m + c <= 2 and m + c >= 1:
                    next_state = State(current.l_m + m, current.l_c + c, 'left')
                    if is_valid(next_state):
                        next_states.append(next_state)
    return next_states

def bfs(start, goal):
    visited = set()
    queue = deque([(start, [])])

    while queue:
        current, path = queue.popleft()
        visited.add(current)

        if current == goal:
            return path + [current]

        for next_state in generate_next_states(current):
            if next_state not in visited:
                queue.append((next_state, path + [current]))
    return None

start_state = State(3, 3, 'left')
goal_state = State(3,3, 'right')
path = bfs(start_state, goal_state)

if path:
    print("Goal state is reachable")
    print("Path to reach the goal state: ")
    for state in path:
        print(state)
else:
    print("Goal state unreachable")
```
