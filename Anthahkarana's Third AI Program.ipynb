```python
class Graph:
    def __init__(self, alist):
        self.alist = alist
        
    def get_neighbors(self, v):
        return self.alist[v]
    
    def h(self, curr):
        H = {
            'A':14,
            'B':12,
            'C':11,
            'D':6,
            'E':4,
            'F':11,
            'G':0
        }
        return H[curr]
    
    def astar(self, start, goal):
        opens = set([start])
        closes = set([])
        cost = {}
        cost[start] = 0
        parent = {}
        parent[start] = start
        
        while len(opens) > 0:
            curr = None
            
            for v in opens:
                if curr is None or cost[v] + self.h(v) < cost[curr] + self.h(curr):
                    curr = v
            if curr is None:
                print('Path does not exist !')
                return None, float('inf')
            
            if curr == goal:
                reconst_path = []
                total_path_cost = cost[curr]
                while parent[curr] != curr:
                    reconst_path.append(curr)
                    curr = parent[curr]
                reconst_path.append(start)
                reconst_path.reverse()
                print('Path found: {}'.format(reconst_path))
                print('Path cost: {}'.format(total_path_cost))
                return reconst_path, total_path_cost
            
            for(neigh, weight) in self.get_neighbors(curr):
                if neigh not in opens and neigh not in closes:
                    opens.add(neigh)
                    parent[neigh] = curr
                    cost[neigh] = cost[curr] + weight
                else:
                    if cost[neigh] > cost[curr] + weight:
                        cost[neigh] = cost[curr] + weight
                        parent[neigh] = curr
                        if neigh in closes:
                            closes.remove(neigh)
                        opens.add(neigh)
            opens.remove(curr)
            closes.add(curr)
        print('Path does not exist !')
        return None, float('inf')
    
alist = {
    'A':[('B', 4), ('C', 3)],
    'B':[('F', 5), ('E', 12)],
    'C':[('D', 7), ('E', 10)],
    'D':[('E', 2)],
    'E':[('G', 5)],
    'F':[('G', 16)]
}

graphics = Graph(alist)
path, cost = graphics.astar('A','G')
```
