```python
class State:
    def __init__(self, j1, j2):
        self.jug1 = j1
        self.jug2 = j2

    def __eq__(self, other):
        return self.jug1 == other.jug1 and self.jug2 == other.jug2

    def __hash__(self):
        return hash((self.jug1, self.jug2))


def dfs(current, visited, path):
    if current == goal:
        return True, path + [current]
    visited.add(current)

    # Fill Jug 1
    if current.jug1 < jug1_capacity:
        next_state = State(jug1_capacity, current.jug2)
        if next_state not in visited:
            found, solution_path = dfs(next_state, visited, path + [current])
            if found:
                return True, solution_path

    # Fill Jug 2
    if current.jug2 < jug2_capacity:
        next_state = State(current.jug1, jug2_capacity)
        if next_state not in visited:
            found, solution_path = dfs(next_state, visited, path + [current])
            if found:
                return True, solution_path

    # Empty Jug 1
    if current.jug1 > 0:
        next_state = State(0, current.jug2)
        if next_state not in visited:
            found, solution_path = dfs(next_state, visited, path + [current])
            if found:
                return True, solution_path

    # Empty Jug 2
    if current.jug2 > 0:
        next_state = State(current.jug1, 0)
        if next_state not in visited:
            found, solution_path = dfs(next_state, visited, path + [current])
            if found:
                return True, solution_path

    # Pour from jug1 to jug 2
    if current.jug1 > 0 and current.jug2 < jug2_capacity:
        pour_amount = min(current.jug1, jug2_capacity - current.jug2)
        next_state = State(current.jug1 - pour_amount, current.jug2 + pour_amount)
        if next_state not in visited:
            found, solution_path = dfs(next_state, visited, path + [current])
            if found:
                return True, solution_path

    # Pour from jug2 to jug1
    if current.jug2 > 0 and current.jug1 < jug1_capacity:
        pour_amount = min(current.jug2, jug1_capacity - current.jug1)
        next_state = State(current.jug1 + pour_amount, current.jug2 - pour_amount)
        if next_state not in visited:
            found, solution_path = dfs(next_state, visited, path + [current])
            if found:
                return True, solution_path

    return False


jug1_capacity = 4
jug2_capacity = 3
goal = State(2, 0)
start = State(4, 3)
visited = set()
found, solution_path = dfs(start, visited, [])
if found:
    print('Goal state is reached')
    print('Path to reach the goal state is as follows')
    for state in solution_path:
        print(state.jug1, state.jug2)
else:
    print('Goal state is not reachable')
```

```python
from collections import deque

#BFS Function to find the optimal path

def bfs(start, end, jug1_cap, jug2_cap):
    queue = deque([(0,0)])
    visited = set([(0,0)])
    parent = {}
    
    while queue:
        current = queue.popleft()
        jug1, jug2 = current
        
        if jug1==end[0] and jug2==end[1]:
            path=[]
            while current != start:
                path.append(current)
                current = parent[current]
            path.append(start)
            return path[::-1]
        
        #Fill Jug 1
        if (jug1 < jug1_cap) and ((jug1_cap, jug2) not in visited):
            queue.append((jug1_cap, jug2))
            parent[(jug1_cap, jug2)] = current
            visited.add((jug1_cap, jug2))
            
        #Fill Jug 2
        if (jug2 < jug2_cap) and ((jug1, jug2_cap) not in visited):
            queue.append((jug1, jug2_cap))
            parent[(jug1, jug2_cap)] = current
            visited.add((jug1, jug2_cap))
            
        #Empty Jug 1
        if jug1 > 0 and (0, jug2) not in visited:
            queue.append((0, jug2))
            parent[(0, jug2)] = current
            visited.add((0, jug2))
        
        #Empty Jug 2
        if jug2 > 0 and (jug1, 0) not in visited:
            queue.append((jug1, 0))
            parent[(jug1, 0)] = current
            visited.add((jug1, 0))
            
        #Pour from jug1 to jug2
        if jug1 > 0:
            pour = min(jug1, jug2_cap - jug2)
            if (jug1 - pour, jug2 + pour) not in visited:
                queue.append((jug1 - pour, jug2 + pour))
                parent[(jug1 - pour, jug2 + pour)] = current
                visited.add((jug1 - pour, jug2 + pour))
                
        #Pour from jug2 to jug1
        if jug2 > 0:
            pour = min(jug2, jug1_cap - jug1)
            if (jug1 + pour, jug2 - pour) not in visited:
                queue.append((jug1 + pour, jug2 - pour))
                parent[(jug1 + pour, jug2 - pour)] = current
                visited.add((jug1 + pour, jug2 - pour))
                
    return None

    
    #Function to print the solution path
def print_path(path):
    if path:
        for state in path:
            print("{},{}".format(state[0], state[1], end= "->"))
        print("Is the goal state")
    else:
        print("No solution found")
            
#Test the implementation
jug1_cap=4
jug2_cap=3
start=(0,0)
end=(2,0)
path=bfs(start, end, jug1_cap, jug2_cap)
print("Solution path")
print_path(path)
```
